```python

import numpy as np

def policy_plotter(c1, c2, r1, r2, w):
    
    result = []
    x_array = np.linspace(0,1,100)
    
    for x in x_array:
        y = min(max(w * abs((x-c1)/r1)**3 + (1-w) * abs((x-c2)/r2)**3,0),1)
        result.append(y)
        
    return result



import matplotlib.pyplot as plt

plt.figure(figsize=(10,20))

selection1 = emadata.loc[emadata.iloc[:,5] < -0.83]

selection2 = emadata.loc[emadata.iloc[:,6] < 14]

selection3 = emadata.loc[emadata.iloc[:,7] < 0.13]

selection4 = emadata.loc[emadata.iloc[:,8] < 450]  

selection5 = emadata.loc[emadata.iloc[:,9] < 2200]

import random
blues = ['#0d5f67', '#0a4c52','#07393d','#052629', '#3d7e85',
        '#6d9fa3', '#9ebfc2', '#cedfe0', '#021314']
browns = ['#7e2e2e', '#642424', '#4b1b1b', '#321212', '#190909',
        '#975757', '#b18181', '#cbabab']
greens = ['#6C9E12', '#567e0e', '#405e0a', '#2b3f07', '#151f03',
         '#89b141', '#a6c470', '#c4d8a0']
purples = ['#69178C', '#541270', '#3f0d54', '#2a0938', '#15041c',
         '#8745a3', '#a573ba', '#c3a2d1']
yellows = ['#DE9733', '#b17828', '#855a1e', '#583c14', '#2c1e0a',
          '#e4ab5b', '#ebc084', '#f1d5ad']

plt.subplot(511)

for index, row in selection1.iterrows():
    plt.plot(np.linspace(0,1,100), policy_plotter(row['c1'],  row['c2'],  row['r1'],  row['r2'],  row['w']), color=random.choice(blues))
plt.title('Policies with effectivness > 0.88')

plt.xlabel('Level of Uncertainty in the System')
plt.ylabel('Probability of Choosing an Explorative Action')

plt.subplot(512)
for index, row in selection2.iterrows():
    plt.plot(np.linspace(0,1,100), policy_plotter(row['c1'],  row['c2'],  row['r1'],  row['r2'],  row['w']), color=random.choice(browns))
plt.title('Policies with containment within 6 weeks')
plt.xlabel('Level of Uncertainty in the System')
plt.ylabel('Probability of Choosing an Explorative Action')

plt.subplot(513)
for index, row in selection3.iterrows():
    plt.plot(np.linspace(0,1,100), policy_plotter(row['c1'],  row['c2'],  row['r1'],  row['r2'],  row['w']), color=random.choice(greens))
plt.title('Policies with difference in met demand < 0.04')
plt.xlabel('Level of Uncertainty in the System')
plt.ylabel('Probability of Choosing an Explorative Action')

plt.subplot(514)
for index, row in selection4.iterrows():
    plt.plot(np.linspace(0,1,100), policy_plotter(row['c1'],  row['c2'],  row['r1'],  row['r2'],  row['w']), color=random.choice(purples))
plt.title('Policies with difference in arrival times < 350')
plt.xlabel('Level of Uncertainty in the System')
plt.ylabel('Probability of Choosing an Explorative Action')

plt.subplot(515)
for index, row in selection5.iterrows():
    plt.plot(np.linspace(0,1,100), policy_plotter(row['c1'],  row['c2'],  row['r1'],  row['r2'],  row['w']), color=random.choice(yellows))
plt.title('Policies with cost per death prevented < 1200')    
plt.xlabel('Level of Uncertainty in the System')
plt.ylabel('Probability of Choosing an Explorative Action')

plt.tight_layout()
plt.show()
```
